# Markov Chain 1:
Learning probabilities letter by letter. Each poem is a different text sample

## Learning probs

```python
probs = {}
Inits = []
order = 5
for poem in data["Poems"]:
    n = len(poem)
    Inits.extend([poem[0:order]])
    for i in range(0,n-order):
        ngram = poem[i:i+order];
        nchar = poem[i+order];
        try:
            probs[ngram] += nchar
        except KeyError:
            probs[ngram] = nchar
```

## Generating a Haiku

### Trial 1

```python
def getHaiku():
    ngram = np.random.choice(Inits)
    poem = ngram
    nchar = 200
    
    for i in range(0,nchar):
        try:
            ch = np.random.choice(list(probs[ngram]))
        except KeyError:
            continue
        else:
            poem = poem + ch
            ngram = poem[-order:]
        
    return poem
```

```python
print(getHaiku())
```

# Markov Chain 2:
Learning as a whole

```python
probs = {}
allPoem = ''.join( data["Poems"])
order = 5
n = len(allPoem)
for i in range(0,n-order):
    ngram = allPoem[i:i+order];
    nchar = allPoem[i+order];
    try:
        probs[ngram] += nchar
    except KeyError:
        probs[ngram] = nchar
```
